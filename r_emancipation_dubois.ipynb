### 5. Creating a Time Series Line Graph

* **Year** is a time variable containing a decade from 1790 to 1870..
* **percent** containts percent of Black U.S. residents who were enslaved during each decade.
* **label** contains a text label for the percent of Black U.S. residents who were free in each decade.
* **label position** contains a value that we will use later to tell R where to put the labels for % free on our graph.

Before recreating Du Bois' **stacked area graph**, we will produce the **time series line graph** that it builds off of. To do this, we will:

Like Du Bois and most time-series graphs, we want to graph the **time variable (year)** on the **x axis**. You will need to edit the code below to graph the correct variable on the **y axis**.
